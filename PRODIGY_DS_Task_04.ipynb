### Analyze and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.

### Generate Dataset

```python
import pandas as pd
import numpy as np

# Generate synthetic data
np.random.seed(0)

# Example dataset columns: 'text' and 'sentiment'
texts = [
    "I love this product! It's amazing.",
    "Terrible service, would not recommend.",
    "Neutral opinion about the company.",
    "Best experience ever, highly recommend!",
    "Disappointed with the product quality.",
    "Great customer support, very helpful.",
    "Not sure about this brand, mixed feelings."
]

sentiments = ['positive', 'negative', 'neutral', 'positive', 'negative', 'positive', 'neutral']

# Create a DataFrame
data = pd.DataFrame({'text': texts, 'sentiment': sentiments})

# Display the dataset
data
```

### Sentiment Analysis

```python
# Install necessary libraries
!pip install nltk
import nltk
nltk.download('vader_lexicon')

# Import libraries
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment_scores(text):
    scores = sid.polarity_scores(text)
    return scores['compound']

# Apply sentiment analyzer to the dataset
data['sentiment_score'] = data['text'].apply(get_sentiment_scores)

# Display dataset with sentiment scores
data
```

### Visualization of Sentiment Patterns

```python
import matplotlib.pyplot as plt

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
colors = {'positive': 'green', 'negative': 'red', 'neutral': 'blue'}
data['sentiment'].value_counts().sort_index().plot(kind='bar', color=data['sentiment'].map(colors))
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Plot sentiment scores distribution
plt.figure(figsize=(8, 6))
data['sentiment_score'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Sentiment Score Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Count')
plt.show()
```
